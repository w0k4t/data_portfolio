```python
import pandas as pd
file_path = '/content/drive/My Drive/data_files/sample_sales_dataset.csv'

df = pd.read_csv(file_path)

display(df.head())
display(df.corr().style.background_gradient(cmap='coolwarm'))
```

```python
import statsmodels.api as sm

# Define the independent variables (X) and the dependent variable (y)
X = df[['Advertising Budget','Social Media Engagement','Website Traffic']]
y = df['Sales']

# Add a constant to the independent variables (intercept term)
X = sm.add_constant(X)

# Create the OLS model
model = sm.OLS(y, X)

# Fit the model
results = model.fit()

# Print the regression results
print(results.summary())
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Sales'], df['Advertising Budget'], df['Social Media Engagement'])
ax.set_xlabel('Sales')
ax.set_ylabel('Advertising Budget')
ax.set_zlabel('Social Media Engagement')

plt.show()
```
